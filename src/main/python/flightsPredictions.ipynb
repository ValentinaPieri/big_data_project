```python
print("Number of Flights by Airline in 2019 dataset")
origin_counts_2019 = df_2019.groupby('Origin').size().reset_index(name='Count').sort_values(by='Count', ascending=False)
print(origin_counts_2019)
```

```python
print("Number of Flights by Origin in 2020 dataset")
origin_counts_2020 = df_2020.groupby('Origin').size().reset_index(name='Count').sort_values(by='Count', ascending=False)
print(origin_counts_2020)
```

```python
print("Number of Flights by Origin in 2021 dataset")
origin_counts_2021 = df_2021.groupby('Origin').size().reset_index(name='Count').sort_values(by='Count', ascending=False)
print(origin_counts_2021)
```

```python
print("Number of Flights by Origin in 2022 dataset")
origin_counts_2022 = df_2022.groupby('Origin').size().reset_index(name='Count').sort_values(by='Count', ascending=False)
print(origin_counts_2022)
```

```python
def cancelled_flights_pie(df):
    cancelled_flights = df['Cancelled'].value_counts()

    plt.figure(figsize=(5, 5))
    cancelled_flights.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90)
    plt.ylabel('')
    plt.show()
```

```python
print('Cancelled Flights in 2019 dataset')
cancelled_flights_pie(df_2019)
print('Cancelled Flights in 2020 dataset')
cancelled_flights_pie(df_2020)
print('Cancelled Flights in 2021 dataset')
cancelled_flights_pie(df_2021)
print('Cancelled Flights in 2022 dataset')
cancelled_flights_pie(df_2022)
```

```python
def diverted_flights_pie(df):
    diverted_flights = df['Diverted'].value_counts()

    plt.figure(figsize=(5, 5))
    diverted_flights.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90)
    plt.ylabel('')
    plt.show()
```

```python
print('Diverted Flights in 2019 dataset')
diverted_flights_pie(df_2019)
print('Diverted Flights in 2020 dataset')
diverted_flights_pie(df_2020)
print('Diverted Flights in 2021 dataset')
diverted_flights_pie(df_2021)
print('Diverted Flights in 2022 dataset')
diverted_flights_pie(df_2022)
```

```python
print("Cancelled and Diverted Flights Rates in 2019 dataset")
rates_2019 = df_2019.groupby('Airline')[['Cancelled','Diverted']].mean() * 100
rates_2019.sort_values('Cancelled', ascending=False)
```

```python
print("Cancelled and Diverted Flights Rates in 2020 dataset")
rates_2020 = df_2020.groupby('Airline')[['Cancelled','Diverted']].mean() * 100
rates_2020.sort_values('Cancelled', ascending=False)
```

```python
print("Cancelled and Diverted Flights Rates in 2021 dataset")
rates_2021 = df_2021.groupby('Airline')[['Cancelled','Diverted']].mean() * 100
rates_2021.sort_values('Cancelled', ascending=False)
```

```python
print("Cancelled and Diverted Flights Rates in 2022 dataset")
rates_2022 = df_2022.groupby('Airline')[['Cancelled','Diverted']].mean() * 100
rates_2022.sort_values('Cancelled', ascending=False)
```

```python
print("Average Departure and Arrival Delay by Airline in 2019 dataset")
delay_by_airline_2019 = df_2019.groupby('Airline')[['DepDelayMinutes','ArrDelayMinutes']].mean()
delay_by_airline_2019.sort_values('DepDelayMinutes', ascending=False)
```

```python
print("Average Departure and Arrival Delay by Airline in 2020 dataset")
delay_by_airline_2020 = df_2020.groupby('Airline')[['DepDelayMinutes','ArrDelayMinutes']].mean()
delay_by_airline_2020.sort_values('DepDelayMinutes', ascending=False)
```

```python
print("Average Departure and Arrival Delay by Airline in 2021 dataset")
delay_by_airline_2021 = df_2021.groupby('Airline')[['DepDelayMinutes','ArrDelayMinutes']].mean()
delay_by_airline_2021.sort_values('DepDelayMinutes', ascending=False)
```

```python
print("Average Departure and Arrival Delay by Airline in 2022 dataset")
delay_by_airline_2022 = df_2022.groupby('Airline')[['DepDelayMinutes','ArrDelayMinutes']].mean()
delay_by_airline_2022.sort_values('DepDelayMinutes', ascending=False)
```

```python
plt.figure(figsize=(12, 8))
dow_2019 = df_2019.groupby('Year')[['DepDelay']].mean()
dow_2020 = df_2020.groupby('Year')[['DepDelay']].mean()
dow_2021 = df_2021.groupby('Year')[['DepDelay']].mean()
dow_2022 = df_2022.groupby('Year')[['DepDelay']].mean()

plt.bar(dow_2019.index, dow_2019['DepDelay'], width=0.2, label='2019', color='lightgreen')
plt.bar(dow_2020.index, dow_2020['DepDelay'], width=0.2, label='2020', color='salmon')
plt.bar(dow_2021.index, dow_2021['DepDelay'], width=0.2, label='2021', color='skyblue')
plt.bar(dow_2022.index, dow_2022['DepDelay'], width=0.2, label='2022', color='gold')

plt.title('Average Departure Delay by Year (2019, 2020, 2021, 2022)')
plt.xlabel('Year')
plt.ylabel('Average Departure Delay (minutes)')
plt.xticks(ticks=[2019, 2020, 2021, 2022], labels=['2019', '2020', '2021', '2022'])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
dow_2019 = df_2019.groupby('Month')[['DepDelay']].mean()
dow_2020 = df_2020.groupby('Month')[['DepDelay']].mean()
dow_2021 = df_2021.groupby('Month')[['DepDelay']].mean()
dow_2022 = df_2022.groupby('Month')[['DepDelay']].mean()

plt.bar(dow_2019.index - 0.3, dow_2019['DepDelay'], width=0.2, label='2019', color='lightgreen')
plt.bar(dow_2020.index - 0.1, dow_2020['DepDelay'], width=0.2, label='2020', color='salmon')
plt.bar(dow_2021.index + 0.1, dow_2021['DepDelay'], width=0.2, label='2021', color='skyblue')
plt.bar(dow_2022.index + 0.3, dow_2022['DepDelay'], width=0.2, label='2022', color='gold')

plt.title('Average Departure Delay by Month (2019, 2020, 2021, 2022)')
plt.xlabel('Month')
plt.ylabel('Average Departure Delay (minutes)')
plt.xticks(dow_2019.index)
plt.legend()
plt.show()
```
